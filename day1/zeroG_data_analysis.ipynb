```python
import numpy as np
from matplotlib import pyplot as plt

# Magic function to make matplotlib inline; other style specs must come AFTER
# %matplotlib inline
%matplotlib notebook
```

```python
file = open("3rd_5.txt", "r") 
lin_accel_x = []
lin_accel_y = []
lin_accel_z = []
lin_accel_mag = []
indexes = []

for i,line in enumerate(file):
    dat = line.split(' ')
    if (dat[0] == "N1"):
        accel = dat[3].split(',')
        a_x = float(accel[0])
        a_y = float(accel[1])
        a_z = float(accel[2])
        indexes.append(i)
        lin_accel_x.append(a_x)
        lin_accel_y.append(a_y)
        lin_accel_z.append(a_z)
        lin_accel_mag.append(np.sqrt(a_x**2+a_y**2+a_z**2))

lin_accel_x = np.array(lin_accel_x)
lin_accel_y = np.array(lin_accel_y)
lin_accel_z = np.array(lin_accel_z)
lin_accel_mag = np.array(lin_accel_mag)
indexes = np.array(indexes)

print(lin_accel_x[:10])
```
